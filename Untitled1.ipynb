```python
size = 20
F = [0 for _ in range(size)]
ctr = 0

def fibonacci(n):
  global ctr
  ctr += 1
  print(str(ctr) + "回目の呼び出し")

  if (n == 0 or n == 1):
    F[n] = 1
    return 1
  if F[n-2] == 0:
    F[n-2] = fibonacci(n-2)
  if F[n-1] == 0:
    F[n-1] = fibonacci(n-1)

  F[n] = F[n-1] + F[n-2]

  print("F[%d] = %d" % (n, F[n]))
  return F[n]

v = fibonacci(size -1)
print("")
print("F[%d] = %d" % (size - 1, v))
```

```python
X = [3,14,6,9]
s = 12
n =len(X)

Y = [None for _ in range(n)]

def BT_subsetsum(level):
  global Y

  print(f'現在の選択:{Y}')

  if level == n:
    total = 0
    for i in range(n):
      total = total + Y[i]*X[i]
    if total == s:
      print(f"sと等しい選び方は{Y}")
    exit()

  else:
    Y[level] = 0
    BT_subsetsum(level+1)
    Y[level] = 1
    BT_subsetsum(level+1)
    Y[level] = None

  if level == 0:
    print(f'sに等しい選び方が存在しない')

BT_subsetsum(0)
```

```python
N = 8
def check(x, col):
  for i , row in enumerate(reversed(col)):
    if (x+1+i == row) or (x-i-1 == row):
      return False
  return True

def search(col):
  if len(col) == N:
    print(col)
    return

  for i in range(N):
    if i not in col:
      if check(i, col):
        col.append(i)
        search(col)
        col.pop()

search([])
```
