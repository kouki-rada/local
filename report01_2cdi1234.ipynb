### 2.1
$x \in \mathbb{R}$„ÇíÂå∫Èñì$[-3, 3]$‰∏ä„ÅÆ‰∏ÄÊßòÂàÜÂ∏É„Å´Âæì„ÅÜÁ¢∫ÁéáÂ§âÊï∞„Å®„ÅóÔºåÁ¢∫ÁéáÂ§âÊï∞$y$„ÅÆÂàÜÂ∏É„Åå$p(y|x) = \mathcal{N}(y | 1.5 x - 0.5, 0.25)$„Å®„Åó„Å¶‰∏é„Åà„Çâ„Çå„Çã„ÇÇ„ÅÆ„Å®„Åô„ÇãÔºé
„Åü„Å†„ÅóÔºå$$ \mathcal{N}(x | \mu, \sigma^2) = \frac{1}{\sqrt{2\pi \sigma^2}} \exp\left(-\frac{(x-\mu)^2}{2\sigma^2}\right)$$„Åß„ÅÇ„ÇãÔºé

„Åì„ÅÆ„Å®„ÅçÔºå‰ª•‰∏ã„ÅÆÊù°‰ª∂„ÇíÊ∫Ä„Åü„Åô„Éó„É≠„Ç∞„É©„É†„Çí‰ΩúÊàê„Åõ„ÇàÔºé

Êù°‰ª∂:
- $x$„ÅÆÂÄ§„Çí100ÂÄãÁîüÊàê„ÅóÔºå„Åù„Çå„Çí1„Å§„ÅÆNumPyÈÖçÂàó„Å´Ê†ºÁ¥ç„Åô„ÇãÔºé
- ÁîüÊàê„Åó„Åü$x$„ÅÆ„Åù„Çå„Åû„Çå„Å´ÂØæÂøú„Åô„Çã$y$„ÇíÁîüÊàê„ÅóÔºå„Åù„Çå„Çí1„Å§„ÅÆNumPyÈÖçÂàó„Å´Ê†ºÁ¥ç„Åô„ÇãÔºé
- ÁîüÊàê„Åó„Åü$x, y$„ÅÆÊï£Â∏ÉÂõ≥„Çí„Éó„É≠„ÉÉ„Éà„Åô„ÇãÔºé„Åü„Å†„ÅóÔºåÊ®™Ëª∏„Çí$x$ÔºåÁ∏¶Ëª∏„Çí$y$„Å®„Åô„ÇãÔºé

```python
# 2.1„ÅÆÂõûÁ≠î„Çí„Åì„Åì„Å´ÂÖ•Âäõ
y = []
x = []
x.append(np.random.uniform(-3, 3, 100))

for i in x:
  u = 1.5 * i - 0.5
  s = 0.25
  y.append(np.random.normal(u, s))

y = np.array(y)
x = np.array(x)

plt.scatter(x, y)
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

### 2.2
$y$„ÅÆÂàÜÂ∏É„ÇíÊú™Áü•„Å®„Åø„Å™„Åó„Å¶Ôºå2.1„ÅßÁîüÊàê„Åó„Åü$x, y$„ÅÆÂÄ§„Åã„Çâ$y$„ÅÆÂàÜÂ∏É„ÇíÊé®ÂÆö„Åô„Çã„Åì„Å®„ÇíËÄÉ„Åà„ÇãÔºéÂÖ∑‰ΩìÁöÑ„Å´„ÅØÔºå„Éë„É©„É°„Éº„Çø$\theta = (w, w_0)^\top, \ w, w_0 \in \mathbb{R}$ „ÇíÁî®„ÅÑ„Å¶ $y$ „ÅÆÂàÜÂ∏É„Çí $p(y | \theta, x) = \mathcal{N}(y | w x + w_0, \sigma^2)$ „Å®„É¢„Éá„É´Âåñ„ÅóÔºåË™§Â∑ÆÈñ¢Êï∞ $$E(\theta) =  \frac{1}{2}\sum_{n=1}^N \left( y_n - y'_n \right)^2$$„ÅÆÊúÄÂ∞èÂåñ„Å´„Çà„Å£„Å¶„Éë„É©„É°„Éº„Çø $\theta$ „ÇíÊúÄÂ∞§Êé®ÂÆö„Åô„ÇãÔºé„Åì„Åì„ÅßÔºå$N$„ÅØ2.1„ÅßÁîüÊàê„Åó„Åü$y$„ÅÆÂÄãÊï∞„Åß„ÅÇ„ÇäÔºå$y_n$„ÅØ$n$Áï™ÁõÆ $(1 \le n \le N)$ „Å´ÁîüÊàê„Åó„Åü$y$„ÅÆÂÄ§„ÇíË°®„ÅôÔºé„Åæ„ÅüÔºå$y'_n = w x_n + w_0$„Åß„ÅÇ„ÇãÔºé

$\frac{\partial}{\partial \theta} E(\theta) = \boldsymbol{0}$ „ÇíËß£„ÅçÔºå„Åù„ÅÆËß£ $\hat{\theta}$ „Çí$x_n, y_n$„ÅÆÂºè„ÅßË°®„ÅõÔºé

**2.2„ÅÆÂõûÁ≠î„Çí„Åì„Åì„Å´ÂÖ•ÂäõÔºàÁ∑®ÈõÜ„Åô„Çã„Å´„ÅØ„ÉÄ„Éñ„É´„ÇØ„É™„ÉÉ„ÇØÔºåÁ∑®ÈõÜ„ÅåÂÆå‰∫Ü„Åó„Åü„ÇâShift+EnterÔºâ**

$E(\theta)$„Å´$y'_n = w x_n + w_0$„Çí‰ª£ÂÖ•„Åô„Çã„Å®„ÄÅ
$$E(\theta) = \frac{1}{2}\sum_{n=1}^N\left(y_n - ( w x_n + w_0 ) \right)^2$$
„Ç∑„Ç∞„É¢„Ç§„ÉâÈñ¢Êï∞„ÅÆÂæÆÂàÜ„Åå„ÄÅ$\frac{d}{da}œÉ(a) = (1 - œÉ(a))œÉ(a)$„Å®Ë°®Áèæ„Åß„Åç„Çã„Åì„Å®„Åã„Çâ„ÄÅ$\frac{\partial}{\partial \theta} E(\theta)$ „Çí$\frac{\partial}{\partial w} E(\theta)$„Å®$\frac{\partial}{\partial w_0} E(\theta)$„ÅßË°®Áèæ„Åß„Åç„Çã„ÄÇ

$\frac{\partial}{\partial w} E(\theta)$„ÅØ„ÄÅ
$$\frac{\partial}{\partial w} E(\theta) = \frac{1}{2}\sum_{n=1}^N\left(y_n - ( w x_n + w_0 ) \right)^2$$
$$\frac{\partial}{\partial w} E(\theta) = \frac{2}{2}\sum_{n=1}^N\left(-x_n)(y_n - ( w x_n + w_0 ) \right)$$
$$\frac{\partial}{\partial w} E(\theta) = - \sum_{n=1}^N\left(x_n(y_n - ( w x_n + w_0 )) \right)$$

$\frac{\partial}{\partial w_0} E(\theta)$„ÅØ„ÄÅ
$$\frac{\partial}{\partial w_0} E(\theta) = \frac{1}{2}\sum_{n=1}^N\left(y_n - ( w x_n + w_0 ) \right)^2$$
$$\frac{\partial}{\partial w_0} E(\theta) = \frac{2}{2}\sum_{n=1}^N\left(- 1)(y_n - ( w x_n + w_0 ) \right)$$
$$\frac{\partial}{\partial w_0} E(\theta) = - \sum_{n=1}^N\left(y_n - ( w x_n + w_0 ) \right)$$

$\frac{\partial}{\partial \theta} E(\theta) = \boldsymbol{0}$ „ÇíËÄÉ„Åà„Çã„Å®„ÄÅ$\frac{\partial}{\partial w} E(\theta)=0$„Å®$\frac{\partial}{\partial w_0} E(\theta)=0$„ÅßË°®Áèæ„Åß„Åç„Çã„ÄÇ‰ª•Èôç„ÄÅ
$$S_x2 = \sum_{n=1}^N\left( x_n^2\right)$$
$$S_xy = \sum_{n=1}^N\left( x_ny_n\right)$$
$$S_x = \sum_{n=1}^N\left( x_n\right)$$
$$S_y = \sum_{n=1}^N\left( y_n\right)$$„Å®ÁΩÆ„ÅçÊèõ„Åà„Çã„ÄÇ

$$\frac{\partial}{\partial w} E(\theta) = 0$$
$$- \sum_{n=1}^N\left(x_n(y_n - ( w x_n + w_0 )) \right) = 0$$
$$\sum_{n=1}^N\left(x_ny_n\right) = \sum_{n=1}^N\left(w x_n^2\right) +\sum_{n=1}^N\left (w_0x_n\right)$$
$$\sum_{n=1}^N\left(x_ny_n\right) = w\sum_{n=1}^N\left( x_n^2\right) +w_0\sum_{n=1}^N\left (x_n\right)$$
$$S_xy = wS_x2 + w_0S_x$$

$$\frac{\partial}{\partial w_0} E(\theta) = 0$$
$$- \sum_{n=1}^N\left(y_n - ( w x_n + w_0 ) \right) = 0$$
$$\sum_{n=1}^N\left(y_n\right) = \sum_{n=1}^N\left(w x_n\right) +\sum_{n=1}^N\left (- w_0\right)$$
$$\sum_{n=1}^N\left(y_n\right) = w \sum_{n=1}^N\left(x_n\right) +Nw_0$$
$$S_y = wSx + Nw_0$$

$w_0$„Å´„Å§„ÅÑ„Å¶„ÄÅ
$$w_0 = \frac{S_y - w S_x}{N}$$

$w_0$„Çí$S_(xy)$„Å´‰ª£ÂÖ•„Åô„Çã„Å®„ÄÅ
$$S_xy = wS_x2 +\frac{S_y - wS_x}{N}S_x$$
$$w(S_x2 - \frac{(S_x)^2}{N}) = S_xy - \frac{S_xS_y}{N}$$
$$w = \frac{N}{NS_x2 - (S_x)^2}\frac{NS_xy - S_xS_y}{N}$$
$$w = \frac{NS_xy - S_xS_y}{NS_x2 - (S_x)^2}$$

$w_0$„Å´$w$„Çí‰ª£ÂÖ•„Åô„Çã„Å®„ÄÅ
$$w_0 = \frac{1}{N}(S_y - \frac{NS_xy - S_xS_y}{NS_x2 - (S_x)^2}S_x)$$

$\frac{\partial}{\partial \theta} E(\theta) = (\frac{\partial}{\partial w} E(\theta)^{T}, \frac{\partial}{\partial w_0} E(\theta)^{T})$„Å®Ë°®Áèæ„Åß„Åç„Çã„ÅÆ„Åß„ÄÅ

$\hat{\theta} = (w^{T}, w_0^{T})$

$$\hat{\theta} = \begin{pmatrix}
w = \frac{NS_xy - S_xS_y}{NS_x2 - (S_x)^2} \\
w_0 = \frac{S_y - w S_x}{N}
\end{pmatrix}$$

$$\hat{\theta} = \begin{pmatrix}
\frac{NS_x2 - (S_x)^2}{NS_xy - S_xS_y} \\
\frac{1}{N}(S_y - \frac{NS_x2 - (S_x)^2}{NS_xy - S_xS_y}S_x)
\end{pmatrix}$$

ùëÅ „ÅØ2.1„ÅßÁîüÊàê„Åó„Åüùë¶„ÅÆÂÄãÊï∞„Å™„ÅÆ„Åß„ÄÅ$N = 100$
$$\hat{\theta} = \begin{pmatrix}
\frac{100\sum_{n=1}^N\left( x_n^2\right) - (\sum_{n=1}^N\left( x_n\right))^2}{100\sum_{n=1}^N\left( x_ny_n\right) - \sum_{n=1}^N\left( x_n\right)\sum_{n=1}^N\left( y_n\right)} \\
\frac{1}{100}(\sum_{n=1}^N\left( y_n\right)- \frac{100\sum_{n=1}^N\left( x_n^2\right) - (\sum_{n=1}^N\left( x_n\right))^2}{100\sum_{n=1}^N\left( x_ny_n\right) - \sum_{n=1}^N\left( x_n\right)\sum_{n=1}^N\left( y_n\right)}\sum_{n=1}^N\left( x_n\right))
\end{pmatrix}$$

### 2.3
2.2„ÅßÊ±Ç„ÇÅ„ÅüÂºè„ÇíÁî®„ÅÑ„Å¶Ôºå2.1„ÅßÁîüÊàê„Åó„Åü$x, y$„Åã„Çâ„Éë„É©„É°„Éº„Çø$\theta$„ÅÆÊúÄÂ∞§Êé®ÂÆöÈáè$\hat{\theta}$„ÇíÊ±Ç„ÇÅ„Çã„Éó„É≠„Ç∞„É©„É†„Çí‰ΩúÊàê„Åõ„ÇàÔºé

```python
# 2.3„ÅÆÂõûÁ≠î„Çí„Åì„Åì„Å´ÂÖ•Âäõ
y = []
x = []
x.append(np.random.uniform(-3, 3, 100))


for i in x:
  u = 1.5 * i - 0.5
  s = 0.25
  y.append(np.random.normal(u, s))

y = np.array(y)
x = np.array(x)
N = len(y)

w = (N * np.sum(x*y) - np.sum(x) * np.sum(y)) / (N * np.sum(x**2) - np.sum(x)**2)
w_0 = (np.sum(y) - w * np.sum(x)) / N

print(f"hat_theta={w},{w_0}")
```

### 3.1
$\boldsymbol{x} = (x_0, x_1)^\top \in \mathbb{R}^2$„Å®„ÅóÔºå
Èñ¢Êï∞ $f(\boldsymbol{x}) = x_0^2 + x_1^2 + 4x_0 + 6x_1 + 9$„ÇíËÄÉ„Åà„ÇãÔºé
„Åì„ÅÆ„Å®„ÅçÔºåÂãæÈÖç$\nabla f(\boldsymbol{x})$„ÇíÊ±Ç„ÇÅ„ÇàÔºé

**3.1„ÅÆÂõûÁ≠î„Çí„Åì„Åì„Å´ÂÖ•ÂäõÔºàÁ∑®ÈõÜ„Åô„Çã„Å´„ÅØ„ÉÄ„Éñ„É´„ÇØ„É™„ÉÉ„ÇØÔºåÁ∑®ÈõÜ„ÅåÂÆå‰∫Ü„Åó„Åü„ÇâShift+EnterÔºâ**

$\boldsymbol{x} = (x_0, x_1)^\top \in \mathbb{R}^2$„Å´„Åä„Åë„ÇãÂãæÈÖç„ÅØÔºå


$\nabla f(\boldsymbol{x}) = (\frac{\partial}{\partial x_1} f(\boldsymbol{x}), ... ,\frac{\partial}{\partial x_0} f(\boldsymbol{x}))^{T}$„Å®ÂÆöÁæ©„Åï„Çå„Çã„ÄÇ

Èñ¢Êï∞ $f(\boldsymbol{x}) = x_0^2 + x_1^2 + 4x_0 + 6x_1 + 9$„Å´„Åä„ÅÑ„Å¶„ÄÅ$\frac{\partial}{\partial x_1} f(\boldsymbol{x})$„Å®$\frac{\partial}{\partial x_0} f(\boldsymbol{x})$„ÇíËÄÉ„Åà„Çã„ÄÇ

$$\frac{\partial}{\partial x_1} f(\boldsymbol{x}) = \frac{\partial}{\partial x_1}(x_0^2 + x_1^2 + 4x_0 + 6x_1 + 9)$$
$$\frac{\partial}{\partial x_1} f(\boldsymbol{x}) = 2x_1 + 6$$

$$\frac{\partial}{\partial x_0} f(\boldsymbol{x}) = \frac{\partial}{\partial x_0}(x_0^2 + x_1^2 + 4x_0 + 6x_1 + 9)$$
$$\frac{\partial}{\partial x_0} f(\boldsymbol{x}) = 2x_0 + 4$$

„Åì„ÅÆ„Å®„ÅçÔºåÂãæÈÖç$\nabla f(\boldsymbol{x})$„ÅØ„ÄÅ
$$\begin{pmatrix}
2x_1 + 6 \\

```python
# 3.2„ÅÆÂõûÁ≠î„Çí„Åì„Åì„Å´ÂÖ•Âäõ

def f(x):
  return x[1]**2 + x[0]**2 + 4*x[1] + 6*x[0] + 9

def d_f(x):
  dfx1 = 2*x[0] + 6
  dfx0 = 2*x[1] + 4
  return np.array([dfx1, dfx0])

def saikyu(fnc, value, learn, error, num):
    x = value
    df = fnc(x)
    for i in range(num):
        nx = x - learn * df
        if np.linalg.norm(nx - x) < error:
            break
        x = nx
    return x

value = np.array([0, 0])
learn = 0.0005
error = 1.0e-10
num = 1000

hat_x = saikyu(d_f, value, learn, error, num)

print("x(x1,x0):", hat_x)
print("f(x):", f(hat_x))
```
