## 問題2
### 2.1
$x \in \mathbb{R}$を区間$[-3, 3]$上の一様分布に従う確率変数とし，確率変数$y$の分布が$p(y|x) = \mathcal{N}(y | 1.5 x - 0.5, 0.25)$として与えられるものとする．
ただし，$$ \mathcal{N}(x | \mu, \sigma^2) = \frac{1}{\sqrt{2\pi \sigma^2}} \exp\left(-\frac{(x-\mu)^2}{2\sigma^2}\right)$$である．

このとき，以下の条件を満たすプログラムを作成せよ．

条件:
- $x$の値を100個生成し，それを1つのNumPy配列に格納する．
- 生成した$x$のそれぞれに対応する$y$を生成し，それを1つのNumPy配列に格納する．
- 生成した$x, y$の散布図をプロットする．ただし，横軸を$x$，縦軸を$y$とする．

```python
# 2.1の回答をここに入力
y = []
x = []
x.append(np.random.uniform(-3, 3, 100))

for i in x:
  u = 1.5 * i - 0.5
  s = 0.25
  y.append(np.random.normal(u, s))

y = np.array(y)
x = np.array(x)

plt.scatter(x, y)
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

### 2.2
$y$の分布を未知とみなして，2.1で生成した$x, y$の値から$y$の分布を推定することを考える．具体的には，パラメータ$\theta = (w, w_0)^\top, \ w, w_0 \in \mathbb{R}$ を用いて $y$ の分布を $p(y | \theta, x) = \mathcal{N}(y | w x + w_0, \sigma^2)$ とモデル化し，誤差関数 $$E(\theta) =  \frac{1}{2}\sum_{n=1}^N \left( y_n - y'_n \right)^2$$の最小化によってパラメータ $\theta$ を最尤推定する．ここで，$N$は2.1で生成した$y$の個数であり，$y_n$は$n$番目 $(1 \le n \le N)$ に生成した$y$の値を表す．また，$y'_n = w x_n + w_0$である．

$\frac{\partial}{\partial \theta} E(\theta) = \boldsymbol{0}$ を解き，その解 $\hat{\theta}$ を$x_n, y_n$の式で表せ．

注:
Google Colabのテキストセルで数式を記述するには，LaTeX形式の記法が利用可能です．
詳しくは，[このページ](https://hwb.ecc.u-tokyo.ac.jp/hwb2023/applications/latex/5min/)などを参考にしてください．



**2.2の回答をここに入力（編集するにはダブルクリック，編集が完了したらShift+Enter）**

$E(\theta)$に$y'_n = w x_n + w_0$を代入すると、
$$E(\theta) = \frac{1}{2}\sum_{n=1}^N\left(y_n - ( w x_n + w_0 ) \right)^2$$
シグモイド関数の微分が、$\frac{d}{da}σ(a) = (1 - σ(a))σ(a)$と表現できることから、$\frac{\partial}{\partial \theta} E(\theta)$ を$\frac{\partial}{\partial w} E(\theta)$と$\frac{\partial}{\partial w_0} E(\theta)$で表現できる。

$\frac{\partial}{\partial w} E(\theta)$は、
$$\frac{\partial}{\partial w} E(\theta) = \frac{1}{2}\sum_{n=1}^N\left(y_n - ( w x_n + w_0 ) \right)^2$$
$$\frac{\partial}{\partial w} E(\theta) = \frac{2}{2}\sum_{n=1}^N\left(-x_n)(y_n - ( w x_n + w_0 ) \right)$$
$$\frac{\partial}{\partial w} E(\theta) = - \sum_{n=1}^N\left(x_n(y_n - ( w x_n + w_0 )) \right)$$

$\frac{\partial}{\partial w_0} E(\theta)$は、
$$\frac{\partial}{\partial w_0} E(\theta) = \frac{1}{2}\sum_{n=1}^N\left(y_n - ( w x_n + w_0 ) \right)^2$$
$$\frac{\partial}{\partial w_0} E(\theta) = \frac{2}{2}\sum_{n=1}^N\left(- 1)(y_n - ( w x_n + w_0 ) \right)$$
$$\frac{\partial}{\partial w_0} E(\theta) = - \sum_{n=1}^N\left(y_n - ( w x_n + w_0 ) \right)$$

$\frac{\partial}{\partial \theta} E(\theta) = \boldsymbol{0}$ を考えると、$\frac{\partial}{\partial w} E(\theta)=0$と$\frac{\partial}{\partial w_0} E(\theta)=0$で表現できる。以降、
$$S_x2 = \sum_{n=1}^N\left( x_n^2\right)$$
$$S_xy = \sum_{n=1}^N\left( x_ny_n\right)$$
$$S_x = \sum_{n=1}^N\left( x_n\right)$$
$$S_y = \sum_{n=1}^N\left( y_n\right)$$と置き換える。

$$\frac{\partial}{\partial w} E(\theta) = 0$$
$$- \sum_{n=1}^N\left(x_n(y_n - ( w x_n + w_0 )) \right) = 0$$
$$\sum_{n=1}^N\left(x_ny_n\right) = \sum_{n=1}^N\left(w x_n^2\right) +\sum_{n=1}^N\left (w_0x_n\right)$$
$$\sum_{n=1}^N\left(x_ny_n\right) = w\sum_{n=1}^N\left( x_n^2\right) +w_0\sum_{n=1}^N\left (x_n\right)$$
$$S_xy = wS_x2 + w_0S_x$$

$$\frac{\partial}{\partial w_0} E(\theta) = 0$$
$$- \sum_{n=1}^N\left(y_n - ( w x_n + w_0 ) \right) = 0$$
$$\sum_{n=1}^N\left(y_n\right) = \sum_{n=1}^N\left(w x_n\right) +\sum_{n=1}^N\left (- w_0\right)$$
$$\sum_{n=1}^N\left(y_n\right) = w \sum_{n=1}^N\left(x_n\right) +Nw_0$$
$$S_y = wSx + Nw_0$$

$w_0$について、
$$w_0 = \frac{S_y - w S_x}{N}$$

$w_0$を$S_(xy)$に代入すると、
$$S_xy = wS_x2 +\frac{S_y - wS_x}{N}S_x$$
$$w(S_x2 - \frac{(S_x)^2}{N}) = S_xy - \frac{S_xS_y}{N}$$
$$w = \frac{N}{NS_x2 - (S_x)^2}\frac{NS_xy - S_xS_y}{N}$$
$$w = \frac{NS_xy - S_xS_y}{NS_x2 - (S_x)^2}$$

$w_0$に$w$を代入すると、
$$w_0 = \frac{1}{N}(S_y - \frac{NS_xy - S_xS_y}{NS_x2 - (S_x)^2}S_x)$$

$\frac{\partial}{\partial \theta} E(\theta) = (\frac{\partial}{\partial w} E(\theta)^{T}, \frac{\partial}{\partial w_0} E(\theta)^{T})$と表現できるので、

$\hat{\theta} = (w^{T}, w_0^{T})$

$$\hat{\theta} = \begin{pmatrix}
w = \frac{NS_xy - S_xS_y}{NS_x2 - (S_x)^2} \\
w_0 = \frac{S_y - w S_x}{N}
\end{pmatrix}$$

$$\hat{\theta} = \begin{pmatrix}
\frac{NS_x2 - (S_x)^2}{NS_xy - S_xS_y} \\
\frac{1}{N}(S_y - \frac{NS_x2 - (S_x)^2}{NS_xy - S_xS_y}S_x)
\end{pmatrix}$$

𝑁 は2.1で生成した𝑦の個数なので、$N = 100$
$$\hat{\theta} = \begin{pmatrix}
\frac{100\sum_{n=1}^N\left( x_n^2\right) - (\sum_{n=1}^N\left( x_n\right))^2}{100\sum_{n=1}^N\left( x_ny_n\right) - \sum_{n=1}^N\left( x_n\right)\sum_{n=1}^N\left( y_n\right)} \\
\frac{1}{100}(\sum_{n=1}^N\left( y_n\right)- \frac{100\sum_{n=1}^N\left( x_n^2\right) - (\sum_{n=1}^N\left( x_n\right))^2}{100\sum_{n=1}^N\left( x_ny_n\right) - \sum_{n=1}^N\left( x_n\right)\sum_{n=1}^N\left( y_n\right)}\sum_{n=1}^N\left( x_n\right))
\end{pmatrix}$$

### 2.3
2.2で求めた式を用いて，2.1で生成した$x, y$からパラメータ$\theta$の最尤推定量$\hat{\theta}$を求めるプログラムを作成せよ．

```python
# 2.3の回答をここに入力
y = []
x = []
x.append(np.random.uniform(-3, 3, 100))


for i in x:
  u = 1.5 * i - 0.5
  s = 0.25
  y.append(np.random.normal(u, s))

y = np.array(y)
x = np.array(x)
N = len(y)

w = (N * np.sum(x*y) - np.sum(x) * np.sum(y)) / (N * np.sum(x**2) - np.sum(x)**2)
w_0 = (np.sum(y) - w * np.sum(x)) / N

print(f"hat_theta={w},{w_0}")
```

## 問題3
### 3.1
$\boldsymbol{x} = (x_0, x_1)^\top \in \mathbb{R}^2$とし，
関数 $f(\boldsymbol{x}) = x_0^2 + x_1^2 + 4x_0 + 6x_1 + 9$を考える．
このとき，勾配$\nabla f(\boldsymbol{x})$を求めよ．

**3.1の回答をここに入力（編集するにはダブルクリック，編集が完了したらShift+Enter）**

$\boldsymbol{x} = (x_0, x_1)^\top \in \mathbb{R}^2$における勾配は，


$\nabla f(\boldsymbol{x}) = (\frac{\partial}{\partial x_1} f(\boldsymbol{x}), ... ,\frac{\partial}{\partial x_0} f(\boldsymbol{x}))^{T}$と定義される。

関数 $f(\boldsymbol{x}) = x_0^2 + x_1^2 + 4x_0 + 6x_1 + 9$において、$\frac{\partial}{\partial x_1} f(\boldsymbol{x})$と$\frac{\partial}{\partial x_0} f(\boldsymbol{x})$を考える。

$$\frac{\partial}{\partial x_1} f(\boldsymbol{x}) = \frac{\partial}{\partial x_1}(x_0^2 + x_1^2 + 4x_0 + 6x_1 + 9)$$
$$\frac{\partial}{\partial x_1} f(\boldsymbol{x}) = 2x_1 + 6$$

$$\frac{\partial}{\partial x_0} f(\boldsymbol{x}) = \frac{\partial}{\partial x_0}(x_0^2 + x_1^2 + 4x_0 + 6x_1 + 9)$$
$$\frac{\partial}{\partial x_0} f(\boldsymbol{x}) = 2x_0 + 4$$

このとき，勾配$\nabla f(\boldsymbol{x})$は、
$$\begin{pmatrix}
2x_1 + 6 \\
2x_0 + 4
\end{pmatrix}$$

### 3.2
3.1で計算した勾配に基づき，関数$f(\boldsymbol{x})$を最急降下法によって最小化するプログラムを作成せよ．プログラムの作成にあたっては，得られた解$\hat{\boldsymbol{x}}$とその時の関数値$f(\hat{\boldsymbol{x}})$を表示させるようにすること．$\boldsymbol{x}$の初期値は自由に設定してよい．また，学習率は，適切に設定すること．

```python
# 3.2の回答をここに入力

def f(x):
  return x[1]**2 + x[0]**2 + 4*x[1] + 6*x[0] + 9

def d_f(x):
  dfx1 = 2*x[0] + 6
  dfx0 = 2*x[1] + 4
  return np.array([dfx1, dfx0])

def saikyu(fnc, value, learn, error, num):
    x = value
    df = fnc(x)
    for i in range(num):
        nx = x - learn * df
        if np.linalg.norm(nx - x) < error:
            break
        x = nx
    return x

value = np.array([0, 0])
learn = 0.0005
error = 1.0e-10
num = 1000

hat_x = saikyu(d_f, value, learn, error, num)

print("x(x1,x0):", hat_x)
print("f(x):", f(hat_x))
```

### 3.3
$\nabla f(x) = 0$を解き，大域最小解 $\tilde{\boldsymbol{x}}$ を求めよ．また，求めた大域最小解と，3.2で得られた解 $\hat{\boldsymbol{x}}$ を比較し，$\hat{\boldsymbol{x}}$ が妥当かどうか考察せよ．

**3.3の回答をここに入力（編集するにはダブルクリック，編集が完了したらShift+Enter）**

勾配$\nabla f(\boldsymbol{x})$は、$\begin{pmatrix}
2x_1 + 6 \\
2x_0 + 4
\end{pmatrix}$なので、

$\begin{pmatrix}
2x_1 + 6 \\
2x_0 + 4
\end{pmatrix} = 0$を解く。

$\begin{pmatrix}
2x_1 + 6 \\
2x_0 + 4
\end{pmatrix} = \begin{pmatrix}
0 \\
0
\end{pmatrix}$なので、

$$(2x_1 + 6) = 0$$
$$x_1 = \frac{-6}{2}$$
$$x_1 = -3$$

$$(2x_0 + 4) = 0$$
$$x_0 = \frac{-4}{2}$$
$$x_0 = -2$$

よって$\tilde{\boldsymbol{x}} = \begin{pmatrix}
-3 \\
-2
\end{pmatrix}$

この時、$f(x)$は
$$f(\boldsymbol{x}) = (-2)^2 + (-3)^2 + 4*(-2) + 6*(-3) + 9 = - 4$$

3.2で得られた解𝑥̂は、
$ 𝑥̂ = \begin{pmatrix}
x_1 \\
x_0
\end{pmatrix}=\begin{pmatrix}
-3 \\
-2
\end{pmatrix}$

代入すると
$f(𝑥̂)=-3.9999999999999982$

𝑥̂は計算で求められた解と同値であり、$f(𝑥̂)$もほぼ同じ値とみなせる誤差である。
よって、3.2で得られた解𝑥̂は妥当だと言える。
